```python
import pandas as pd

def load_data_from_github(experiment, ip_address):
    """
    Loads data from a CSV file on GitHub based on experiment and IP address.

    Args:
        experiment (str): The name of the experiment (e.g., 'chuze').
        ip_address (str): The IP address in the format 'x.x.x.x:port'.

    Returns:
        pd.DataFrame: A DataFrame containing the loaded data.
    """
    # Base URL of the GitHub repository
    base_url = "https://raw.githubusercontent.com/CVUT-FS-12110/letni-skola-ai-2025/refs/heads/master/Den02/data/"

    # Replace dots in IP address with underscores
    formatted_ip = ip_address.replace('.', '_')
    formatted_ip = formatted_ip.replace(':', '_')

    # Construct the final URL
    csv_url = f"{base_url}{experiment}_{formatted_ip}.csv"

    # Read the CSV data directly into a pandas DataFrame
    df = pd.read_csv(csv_url)

    # Rename columns
    df.columns = ['time_s', 'acc_x', 'acc_y', 'acc_z', 'acc_abs']
    df['activity'] = experiment

    return df

# Example usage of the function:
experiment_name = "chuze"
phone_ip = "10.0.1.13:8080"

df = load_data_from_github(experiment_name, phone_ip)

print(f"Data successfully loaded for experiment '{experiment_name}' from IP '{phone_ip}'!")
display(df.head())
```

```python
import numpy as np

# ==============================================================================
# ČÁST 1: Definice naší vlastní funkce ("kuchyňský robot")
# ==============================================================================

def nakrajej_data_na_okenka(dataframe, nazev_aktivity, ciselna_nalepka, delka_okenka, krok):
    """
    Tato funkce vezme velkou tabulku, vybere z ní jednu aktivitu
    a "nakrájí" její data na menší, překrývající se okénka.

    Vstupy (ingredience):
      - dataframe: Velká tabulka se všemi daty.
      - nazev_aktivity: Textový název aktivity, kterou chceme zpracovat (např. 'skoky').
      - ciselna_nalepka: Číslo, kterým označíme všechna okénka (např. 0).
      - delka_okenka: Velikost jednoho okénka.
      - krok: O kolik se posuneme pro další okénko.

    Výstupy (hotové jídlo):
      - Dva seznamy: jeden se samotnými okénky a druhý s jejich nálepkami.

    Příklad:
    Data: [1, 2, 3, 4, 5, 6, 7, 8],  delka_okenka=4, krok=2

    Okénko 1: [1, 2, 3, 4]
        Okénko 2:   [3, 4, 5, 6]
            Okénko 3:     [5, 6, 7, 8]
    """
    print(f"--- Zpracovávám aktivitu: {nazev_aktivity} ---")

    # Vybereme si z velké tabulky jen řádky, které patří naší aktivitě
    data_pro_aktivitu = dataframe[dataframe['activity'] == nazev_aktivity]['acc_abs'].values

    # Připravíme si prázdné seznamy pro výsledky
    vysledna_okenka = []
    vysledne_nalepky = []

    # Projdeme data klouzavým oknem
    for i in range(0, len(data_pro_aktivitu) - delka_okenka, krok):
        okenko = data_pro_aktivitu[i : i + delka_okenka]
        vysledna_okenka.append(okenko)
        vysledne_nalepky.append(ciselna_nalepka)

    print(f"Pro '{nazev_aktivity}' bylo vytvořeno {len(vysledna_okenka)} okének.")

    return vysledna_okenka, vysledne_nalepky


# ==============================================================================
# ČÁST 2: Použití naší nové funkce
# ==============================================================================

# Nastavení parametrů pro "krájení"
DELKA_OKENKA = 100
KROK = 10

# Teď zavoláme naši funkci třikrát, pokaždé s jinými "ingrediencemi"
okenka_stani, nalepky_stani = nakrajej_data_na_okenka(master_df, 'klid', 0, DELKA_OKENKA, KROK)
okenka_chuze, nalepky_chuze = nakrajej_data_na_okenka(master_df, 'chuze', 1, DELKA_OKENKA, KROK)
okenka_skakani, nalepky_skakani = nakrajej_data_na_okenka(master_df, 'skoky', 2, DELKA_OKENKA, KROK)

# Spojíme výsledky ze všech tří volání dohromady
X_vsechna_okenka = okenka_stani + okenka_chuze + okenka_skakani
y_vsechny_nalepky = nalepky_stani + nalepky_chuze + nalepky_skakani

# Finální převod na formát pro TensorFlow
X_train = np.array(X_vsechna_okenka)
y_train = np.array(y_vsechny_nalepky)

print("\n\nData byla úspěšně 'nakrájena' na okénka pomocí funkce!")
print("Celkový tvar trénovacích dat (počet oken, délka okna):", X_train.shape)
print("Celkový tvar trénovacích nálepek (počet oken):", y_train.shape)
```

```python
# Importujeme funkci pro snadné rozdělení dat
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical

# Rozdělíme naše "okenka" a "nálepky" na trénovací (80 %) a testovací (20 %) část.
# stratify=y_train zajistí, že v obou sadách bude podobný poměr všech tří aktivit.
X_train, X_test, y_train, y_test = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42, stratify=y_train)

# Převedeme nálepky z čísel (0, 1, 2) na tzv. "one-hot" formát ([1,0,0], [0,1,0], [0,0,1])
# Je to formát, kterému neuronová síť pro klasifikaci rozumí nejlépe.
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)


print("Data byla rozdělena:")
print("Počet okenek pro trénování:", X_train.shape[0])
print("Počet okenek pro testování:", X_test.shape[0])
print("\nUkázka 'one-hot' nálepky pro kategorii 'Chůze' (číslo 1):", y_train_cat[y_train == 1][0])
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout

# Zjistíme si délku jednoho okénka z tvaru našich dat
DELKA_OKENKA = X_train.shape[1]

# Začneme stavět model, vrstvu po vrstvě
model = Sequential([
    # 1. Vstupní vrstva: "Flatten" vezme naše okénko (např. 50 čísel) a "narovná" ho
    # do jedné dlouhé řady, aby tomu další vrstvy rozuměly.
    Flatten(input_shape=(DELKA_OKENKA,)),

    # 2. Skrytá vrstva: Toto je první "chytrá" vrstva. 64 neuronů se bude snažit
    # najít v datech jednoduché vzorce. 'relu' je standardní a osvědčená aktivace.
    Dense(64, activation='relu'),

    # "Dropout" je technika proti "šprtání". Náhodně "vypne" 20 % neuronů,
    # aby se síť nespoléhala jen na pár z nich a lépe zobecňovala.
    Dropout(0.2),

    # 3. Další skrytá vrstva: Neurony v této vrstvě se učí složitější vzorce
    # z toho, co jim připravila předchozí vrstva.
    Dense(32, activation='relu'),

    # 4. Výstupní vrstva: Musí mít přesně 3 neurony - jeden pro každou naši aktivitu.
    # Aktivace 'softmax' na konci zajistí, že součet výstupů bude 1. Můžeme to
    # interpretovat jako pravděpodobnost, např. [0.1, 0.8, 0.1] -> 80% šance, že je to chůze.
    Dense(3, activation='softmax')
])

# Necháme si ukázat přehlednou strukturu našeho modelu
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
optimizer = Adam(learning_rate=0.0005)
# Kompilace modelu
model.compile(optimizer=optimizer, # 'adam' je osvědčený a spolehlivý optimalizátor
              loss='categorical_crossentropy', # Loss funkce měří, jak moc se model plete
              metrics=['accuracy']
              ) # Chceme sledovat přesnost (accuracy)

# Spustíme samotné trénování!
print("\n--- Spouštím trénování modelu ---")
history = model.fit(X_train, y_train_cat,          # Trénovací data
                    epochs=30,                      # Počet "kol", kolikrát se model podívá na všechna data
                    batch_size=8,                  # Po kolika okénkách si model upraví váhy
                    validation_data=(X_test, y_test_cat), # Testovací data pro průběžnou kontrolu
                    verbose=1)                      # Chceme vidět průběh (1=ano, 0=ne)

print("--- Trénování dokončeno! ---")
```

```python
# Vyhodnocení na testovací sadě
test_loss, test_accuracy = model.evaluate(X_test, y_test_cat, verbose=0)
print(f"\nPřesnost modelu na testovacích datech: {test_accuracy * 100:.2f} %")

# Vykreslíme si graf, jak se model učil
plt.figure(figsize=(14, 5))

# Graf přesnosti
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Trénovací přesnost')
plt.plot(history.history['val_accuracy'], label='Validační (testovací) přesnost')
plt.title('Průběh přesnosti během trénování')
plt.xlabel('Epocha')
plt.ylabel('Přesnost')
plt.legend()
plt.grid(True)

# Graf ztráty (chyby)
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Trénovací ztráta')
plt.plot(history.history['val_loss'], label='Validační (testovací) ztráta')
plt.title('Průběh ztráty během trénování')
plt.xlabel('Epocha')
plt.ylabel('Ztráta')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.lines import Line2D

# Klasifikovat chceme celkové zrychleni
data_k_predikci = df_validace['acc_abs']

# Časová osa z validačních dat
casova_osa = df_validace['time_s']

# ==============================================================================
# KROK 2: "Nakrájení" testovacích dat na okénka
# ==============================================================================
print("Připravuji okénka pro predikci...")

# Po jak velkých krocích chci v signálu predikovat způsob pohybu?
krok = 20

okenka_k_predikci = []

for i in range(0, len(data_k_predikci) - DELKA_OKENKA, krok):
    okenko = data_k_predikci[i : i + DELKA_OKENKA]
    okenka_k_predikci.append(okenko)

okenka_k_predikci = np.array(okenka_k_predikci)

# ==============================================================================
# KROK 3: Predikce pomocí natrénovaného modelu
# ==============================================================================
print("Spouštím predikci na testovacích datech...")
predikovane_nalepky = np.argmax(model.predict(okenka_k_predikci, verbose=0), axis=1)
print("Predikce dokončena!")

# ==============================================================================
# KROK 4: Finální vizualizace do jednoho grafu
# ==============================================================================
print("Vykresluji finální graf...")
plt.figure(figsize=(10, 6))

# Vykreslíme původní signál
plt.plot(casova_osa, data_k_predikci, color='black', label='Původní signál', lw=2)

# Připravíme si barvy pro jednotlivé aktivity
colors = {0: 'blue', 1: 'green', 2: 'red'} # 0: Stání, 1: Chůze, 2: Skákání

# Vykreslíme predikce jako barevné úseky pod grafem
for i in range(len(predikovane_nalepky)):
    start_cas = casova_osa[i * krok]
    konec_cas = casova_osa[i * krok + DELKA_OKENKA]
    nalepka = predikovane_nalepky[i]
    plt.axvspan(start_cas, konec_cas, color=colors[nalepka], alpha=0.1)

# Vytvoříme si vlastní legendu, aby byla hezky vidět
legend_elements = [Line2D([0], [0], color='blue', lw=8, alpha=0.4, label='Predikce: Stání'),
                   Line2D([0], [0], color='green', lw=8, alpha=0.4, label='Predikce: Chůze'),
                   Line2D([0], [0], color='red', lw=8, alpha=0.4, label='Predikce: Skákání'),
                   Line2D([0], [0], color='gray', lw=2, label='Původní signál')]

# Nastavíme popisky a název grafu
plt.title('Finální test modelu na souvislých datech', fontsize=18)
plt.xlabel('Čas (s)', fontsize=14)
plt.ylabel('Absolutní zrychlení (m/s^2)', fontsize=14)
plt.legend(handles=legend_elements, fontsize=12)
plt.grid(True)
plt.tight_layout()

# Zobrazíme hotový graf
plt.show()
```
